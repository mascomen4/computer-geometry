```python
from matplotlib import pyplot as plt
from matplotlib import animation as animation
import numpy as np
```

```python
# Reference: https://pythonmatplotlibtips.blogspot.com/2017/12/cycloid-animation-artistanimation.html

# Radius of the circle
r = 3
# Desired number of segments
s = 100



def circle(a, b, r, s):
    """
    Returns all the points with desired center and radius
    :param a: x of the circle
    :param b: y of the circle
    :param r: radius of the circle
    :param s: desired number of segments
    :return: [x], [y] - points of the circle
    """
    x, y = [], []
    for fi in np.linspace(0, 2*np.pi, s):
        x.append(a + r*np.cos(fi))
        y.append(b + r*np.sin(fi))
    return x, y


"""
Draws a cycloid
:param r: radius
:param s: desired number of segments
:return: fig - plt figure, lns - lines.
"""
theta = np.linspace(0, 4*np.pi, s)
lns = []
fig = plt.figure(figsize=(12,4))
plt.axes().set_aspect('equal')
plt.xlim(-6,45)
plt.ylim(0,10)
plt.xlabel('X')
plt.ylabel('Y')
# plt.legend(ncol=2, title='Циклоида', framealpha=0.7, fontsize=12)
plt.grid()
print("Figure configuration success...")

cycloid_x = r * (theta - np.sin(theta))
cycloid_y = r * (1 - np.cos(theta))
cycloid_c = r * theta

for i, fi in enumerate(theta):
    x, y = circle(cycloid_c[i], r, r, s)
    # Draw the circle
    circ, = plt.plot(x, y ,"b-", linewidth=1)
    # Draw it's radius points
    radius, = plt.plot(cycloid_x[i], cycloid_y[i], "bo", linewidth=1)
    # Draw it's center
    center, = plt.plot(cycloid_c[i], r, "bo", linewidth=1)
    # Draw the cycloid
    cycl, = plt.plot(cycloid_x[:i + 1], cycloid_y[:i + 1], "g-", label="Циклоида")
    # Draw the arrow
    if i != (len(theta) - 1):
        point0 = [cycloid_x[i], cycloid_y[i]]
        point1 = [cycloid_x[i+1], cycloid_y[i+1]]
        # Строка заимствована у преподавателя
        style = dict(length_includes_head=True, width=0.4, head_width=.8, color='black')
        #arrow = plt.arrow(*point0, *point1, **style)
    lns.append([circ, center, cycl, radius])

```

```python
ani = animation.ArtistAnimation(fig, lns, interval=50)
print("Animation saving success...")
# ani.save('cycloid_ArtistAnimation.mp4', writer='ffmpeg')
ani.save('cycloid_ArtistAnimation.gif', writer='imagemagick')
ani
```
