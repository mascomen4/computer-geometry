```python
from matplotlib import pyplot as plt
from matplotlib import patches as mpatches
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import numpy as np
```

```python
# Radius of the circle
r = 3
# Desired number of segments
s = 1000

theta = np.linspace(0, 4*np.pi, 1000)
x = r * (theta - np.sin(theta))
y = r * (1 - np.cos(theta))

#      ->   
dx = r * (1 - np.cos(theta))
dy = r * (np.sin(theta))


# Fig config
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(1,1,1)

ax.set_ylabel(f" $y$", fontsize=15)
ax.set_xlabel(f" $x$", fontsize=15)

ax.tick_params(axis='both', which='minor', labelsize=15)
ax.tick_params(axis='both', which='major', labelsize=15)

plt.xticks(np.arange(min(x), max(x)+1, 5.))
plt.yticks(np.arange(min(y), max(y)+1, 5.))

ax.xaxis.set_minor_locator(MultipleLocator(1))

ax.xaxis.set_minor_locator(AutoMinorLocator())

ax.tick_params(which='both', width=1)
ax.tick_params(which='major', length=10)
ax.tick_params(which='minor', length=5)


# 
ax.grid()
#   . 
ax.set_aspect('equal')

circle = mpatches.Circle([10, r], r, color="blue", fill=False, lw=2)

# Plot the cycloid
ax.plot(x, y, color='green', label="", lw=3)
# Plot the circle
ax.add_patch(circle)
# Plot the center of the circle
ax.plot(10, r, 'bo', markersize=8)

#  20   
pdx = dx[::20]
pdy = dy[::20]

#   scale -    . 
ax.quiver(x[::20], y[::20], pdx, pdy, width=.004, minlength=.02, scale=300, label=" ")

ax.legend(ncol=1, framealpha=.7, fontsize=15)
```
