```python
from matplotlib import pyplot as plt
import numpy as np

import matplotlib.patches as mpatches

%matplotlib inline
```

```python
def plot_regular_polygon(xy, n_edges, edge_length):
    """
        Draws a Regular Polygon.
    """

    fig = plt.figure(num=1, figsize=(8,8))
    ax1 = fig.add_subplot(1, 1, 1)
    ax1.set_xlabel(r"Ось O$x$")
    ax1.set_ylabel("Ось O$у$")
    ax1.set_title("Правильный многоугольник")
    ax1.grid()

    # Formula from https://www.google.com/url?sa=i&url=https%3A%2F%2Fuztest.ru%2Fabstracts%2F%3Fidabstract%3D50&psig=AOvVaw28_wgCaIvsB02FonYUgm_p&ust=1631614998772000&source=images&cd=vfe&ved=0CAsQjRxqFwoTCJjMnend-_ICFQAAAAAdAAAAABAJ
    arg = np.radians(180/n_edges)
    out_radius = edge_length/(np.sin(arg) * 2)
    in_radius = out_radius * np.cos(arg)

    # The lower the grid_scale, the closer you see the Polygon.
    grid_scale = 2

    ax1.set_xlim(left=xy[0] - out_radius * grid_scale, right=xy[0] + out_radius * grid_scale)
    ax1.set_ylim(bottom=xy[1] - out_radius * grid_scale, top=xy[1] + out_radius * grid_scale)

    rpolygon = mpatches.RegularPolygon(xy, n_edges, out_radius, fill=False, linewidth=2)
    out_circle = mpatches.Circle(xy, out_radius, fill=False, color="red")
    in_circle = mpatches.Circle(xy, in_radius, fill=False, color='blue')

    path = rpolygon.get_path()
    vert = path.vertices
    trans = rpolygon.get_patch_transform()
    points = trans.transform(vert)

    ax1.add_patch(rpolygon)
    ax1.add_patch(out_circle)
    ax1.add_patch(in_circle)
    ax1.plot(points[:, 0], points[:, 1], 'bo')

    for i in range(len(points[:-1])):
        ax1.text(x=1.2 * points[i, 0], y=1.2 * points[i, 1], s=f"$P_{i}$", verticalalignment='center', horizontalalignment='center',
                 fontstyle="italic", fontsize=20)
```

```python
plot_regular_polygon(np.array([0, 0]), 6, 4)
```
