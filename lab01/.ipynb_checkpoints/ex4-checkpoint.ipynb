```python
from matplotlib import pyplot as plt 
from matplotlib import animation as animation
import numpy as np
import sympy
from IPython.display import display
```

```python
# Вычислим аналитически формулы производных. 

a = 2

t = sympy.symbols("t")
sympy.init_printing(use_unicode=True)
x_sym = a * sympy.cos(t)**3
y_sym = a * sympy.sin(t)**3

diff_x_1 = sympy.diff(x_sym)
diff_x_2 = sympy.diff(diff_x_1)
diff_y_1 = sympy.diff(y_sym)
diff_y_2 = sympy.diff(diff_y_1)

# Выведем красиво производные, чтобы затем записать их в формулы
display(diff_x_1)
display(diff_x_2)
display(diff_y_1)
display(diff_y_2)
```

```python
# Построим график. 
fig = plt.figure(figsize=(6,6))
ax1 = fig.add_subplot(1,1,1)
ax1.grid()

theta = np.linspace(10e-6, 4*np.pi - 10e-6, 1000)
x = a * np.cos(theta)**3
y = a * np.sin(theta)**3

# При изменении параметра а, необходимо вручную переписать производные. 
diffx1 = -6*np.sin(theta)*np.cos(theta)**2
diffx2 = 12*np.sin(theta)**2 * np.cos(theta) - 6*np.cos(theta)**3
diffy1 = 6*np.sin(theta)**2 * np.cos(theta)
diffy2 = -6*np.sin(theta)**3 + 12*np.sin(theta)*np.cos(theta)**2

denominator = diffx1 * diffy2 - diffx2 * diffy1
squared_sum = diffx1**2 + diffy1**2

ev_x = x - ( diffy1 * squared_sum ) / denominator
ev_y = y + ( diffx1 * squared_sum ) / denominator

ax1.plot(x, y)
ax1.plot(ev_x, ev_y)
```
